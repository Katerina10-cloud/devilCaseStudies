```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# Step 1: Load the CSV
df = pd.read_csv('dataset/data.csv')  # Replace with your CSV file path

# Step 2: Prepare the data (assuming columns are 'y', 'mu', 'log_theta', 'll')
features = df[['y', 'mu', 'log_theta']].values
target = df['ll'].values

# Step 3: Standardize the features
feature_scaler = StandardScaler()
features = feature_scaler.fit_transform(features)

#target_scaler = StandardScaler()
#target = target_scaler.fit_transform(target.reshape(-1, 1)).squeeze(-1)

# Step 4: Split data into training and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Step 5: Create a custom Dataset
class MyDataset(Dataset):
    def __init__(self, features, target):
        self.features = torch.tensor(features, dtype=torch.float32)
        self.target = torch.tensor(target, dtype=torch.float32)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return self.features[idx], self.target[idx]

# Step 6: Create DataLoader for both train and test sets
train_dataset = MyDataset(X_train, y_train)
train_dataloader = DataLoader(train_dataset, batch_size=512, shuffle=True)

test_dataset = MyDataset(X_test, y_test)
test_dataloader = DataLoader(test_dataset, batch_size=1024, shuffle=False)
```

```python
# Step 7: Define the Neural Network architecture
class NeuralNetwork(nn.Module):
    def __init__(self, input_dim=3, output_dim=1, hidden_layers=[64, 32]):
        """
        Initialize the neural network.
        
        Parameters:
        - input_dim: Number of input features (default is 3)
        - output_dim: Number of output features (default is 1)
        - hidden_layers: List specifying the number of neurons in each hidden layer
        """
        super(NeuralNetwork, self).__init__()

        # Create a list of layers (input layer, hidden layers, output layer)
        layers = []

        # Input layer (3 features -> first hidden layer)
        layers.append(nn.Linear(input_dim, hidden_layers[0]))
        
        # Hidden layers
        for i in range(1, len(hidden_layers)):
            layers.append(nn.ReLU())
            layers.append(nn.Linear(hidden_layers[i-1], hidden_layers[i]))

        # Output layer (final hidden layer -> output)
        layers.append(nn.ReLU())
        layers.append(nn.Linear(hidden_layers[-1], output_dim))

        # Define the model as a sequential list of layers
        self.model = nn.Sequential(*layers)

    def forward(self, x):
        return self.model(x)

# Step 8: Initialize the model, loss function, and optimizer
model = NeuralNetwork(hidden_layers=[64, 32, 16, 8])
criterion = nn.MSELoss()  # Mean Squared Error loss for regression
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Step 9: Training the model and recording the loss
num_epochs = 2000
train_losses = []
test_losses = []

for epoch in range(num_epochs):
    model.train()
    running_train_loss = 0.0
    
    # Training loop
    for inputs, labels in train_dataloader:
        optimizer.zero_grad()
        
        # Forward pass
        outputs = model(inputs)
        
        # Compute training loss
        loss = criterion(outputs.squeeze(), labels)
        
        # Backward pass
        loss.backward()
        
        # Optimize
        optimizer.step()
        
        running_train_loss += loss.item()
    
    avg_train_loss = running_train_loss / len(train_dataloader)
    train_losses.append(avg_train_loss)  # Store the average loss for this epoch
    
    # Compute test loss
    model.eval()
    running_test_loss = 0.0
    with torch.no_grad():
        for inputs, labels in test_dataloader:
            # Forward pass
            outputs = model(inputs)
            
            # Compute test loss
            loss = criterion(outputs.squeeze(), labels)
            running_test_loss += loss.item()
    
    avg_test_loss = running_test_loss / len(test_dataloader)
    test_losses.append(avg_test_loss)
    
    # Print losses every 100 epochs
    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {avg_train_loss:.10f}, Test Loss: {avg_test_loss:.10f}')
```

```python
# Step 10: Plotting the training and test loss
import matplotlib.pyplot as plt

plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.yscale('log')
plt.legend()
plt.title('Train and Test Loss Over Epochs')
plt.show()
```
